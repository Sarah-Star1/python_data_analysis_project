# How In demand skills trending for data analyst

### Methodology:
1. aggregate skills count monthly.
2. reanalyze based on the percentage of total jobs.
3. plot the monthly skill demand.

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
from datasets import load_dataset

# Load the dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Convert the 'skills' column from string representation of list to actual list
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

#cleaning the data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#Filtering skills for data analysts in the US in 2023:
df_da_us = df[(df['job_title_short'] == 'Data Analyst') &  (df['job_country'] == 'United States')]
df_da_us = df_da_us[df_da_us['job_posted_date'].dt.year == 2023]

df_da_us_exploded = df_da_us.explode('job_skills')
df_da_us_exploded['job_posted_month'] = pd.to_datetime(df_da_us_exploded['job_posted_date']).dt.strftime('%b')

x = df_da_us_exploded.groupby(['job_skills','job_posted_month']).size().sort_values(ascending=False).to_frame()
x.reset_index(inplace=True)
```

```python
my_skills= ['python','SQL',]
x.pivot_table(index='job_posted_month', columns= ['job_skills'].isin(my_skills), aggfunc='median')
```
